```python
df_final['date_time'] = pd.to_datetime(df_final['date_time'])
```

```python
df_final = df_final.sort_values(by=['client_id', 'visit_id', 'date_time'])
```

```python
df_final
```

```python

unique_steps = df['process_step'].unique()
print(unique_steps)
```

```python
df_final['time_spent'] = df_final.groupby('visit_id')['date_time'].diff()
```

```python
# Drop NaT values from time_spent and filter out 'start' steps
df_time_spent = df_final.dropna(subset=['time_spent'])
df_time_spent = df_time_spent[df_time_spent['process_step'] != 'start']
```

```python

df_time_spent['time_spent_minutes'] = df_time_spent['time_spent'].dt.total_seconds() / 60
```

```python
unique_steps = df_time_spent['process_step'].unique()
print("Unique process steps:", unique_steps)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='process_step', y='time_spent_minutes', data=df_time_spent, palette='coolwarm')
plt.title('Distribution of Time Spent on Each Process Step (in minutes)')
plt.xlabel('Process Step')
plt.ylabel('Time Spent (minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='process_step', y='time_spent_minutes', data=df_time_spent, palette='coolwarm')
plt.title('Distribution of Time Spent on Each Process Step (in minutes)')
plt.xlabel('Process Step')
plt.ylabel('Time Spent (minutes)')
plt.xticks(rotation=45)
plt.ylim(0, 200)  # Set y-axis limit to a maximum of 200 minutes
plt.show()
```

```python
import matplotlib.pyplot as plt

# Set up a 2x2 grid for the four histograms
plt.figure(figsize=(14, 10))

# Loop over each unique process step and create a histogram
for i, step in enumerate(['step_1', 'step_2', 'step_3', 'confirm'], 1):
    plt.subplot(2, 2, i)  # 2x2 grid
    step_data = df_time_spent[df_time_spent['process_step'] == step]
    
    plt.hist(step_data['time_spent_minutes'], bins=30, color='teal', edgecolor='black')
    plt.title(f'Distribution of Time Spent on {step} (in minutes)')
    plt.xlabel('Time Spent (minutes)')
    plt.ylabel('Frequency')
    plt.xlim(0, 200)  # Limit x-axis to a maximum of 200 minutes for clearer visualization

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Set up a 2x2 grid for the four histograms
plt.figure(figsize=(14, 10))

# Loop over each unique process step and create a histogram with a log scale on the y-axis
for i, step in enumerate(['step_1', 'step_2', 'step_3', 'confirm'], 1):
    plt.subplot(2, 2, i)  # 2x2 grid
    step_data = df_time_spent[df_time_spent['process_step'] == step]
    
    plt.hist(step_data['time_spent_minutes'], bins=30, color='teal', edgecolor='black')
    plt.title(f'Distribution of Time Spent on {step} (in minutes)')
    plt.xlabel('Time Spent (minutes)')
    plt.ylabel('Frequency')
    plt.xlim(0, 200)  # Limit x-axis to a maximum of 200 minutes for clearer visualization
    plt.yscale('log')  # Set the y-axis to logarithmic scale

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(14, 10))
gs = gridspec.GridSpec(2, 2, figure=fig)

steps = ['step_1', 'step_2', 'step_3', 'confirm']

# Create each subplot with centered configuration
for i, step in enumerate(steps):
    row = i // 2
    col = i % 2
    ax = fig.add_subplot(gs[row, col])  # Place each plot in a 2x2 grid
    step_data = df_time_spent[df_time_spent['process_step'] == step]
    
    sns.histplot(step_data['time_spent_minutes'], bins=30, color='teal', edgecolor='black', ax=ax)
    ax.set_title(f'Distribution of Time Spent on {step} (in minutes)')
    ax.set_xlabel('Time Spent (minutes)')
    ax.set_ylabel('Frequency')
    ax.set_xlim(0, 200)
    ax.set_yscale('log')  # Logarithmic scale on the y-axis

plt.tight_layout(pad=3.0)  # More padding for better spacing
plt.show()
```

```python
avg_time_per_client_step = df_time_spent.groupby(['client_id', 'process_step'])['time_spent_minutes'].mean()

avg_time_per_client_step = avg_time_per_client_step.reset_index()

print(avg_time_per_client_step)
```
